```python
#!/usr/bin/env python3# R441v6_Hotco_Walnut_Finder_NoSkip# Previous version: R441v5 - processed Bread with skipping# cchat: current conversationscript_name = "R441v6_Hotco_Walnut_Finder_NoSkip"v_version = 6"""PURPOSE: Find ALL hotspot-coldspot pairs in Walnut dataset - NO SKIPSDataset: Walnut - 4DNFI3ULPBI5.pairs.gz (H1-hESC stem cells Micro-C, 29GB)Method: Hotspots with 5x5 window filter, coldspots at 0-bin offset checking 2x20 positionsProcessing: NO SKIPS - process entire file continuouslyChanges v6: Updated to Walnut dataset (Spock codename), no skipping for 29GB file"""import os, sys, time, gzip, picklefrom datetime import datetime, timezone, timedeltaimport numpy as npimport matplotlib.pyplot as pltimport matplotlib.colors as mcolorsfrom collections import defaultdict# Nice the processtry:    os.nice(10)except Exception:    pass# DATASET CONFIGURATION - Use variables for easy swappingDATASET_NAME = "Walnut"DATASET_CODENAME = "Spock"DATASET_FILE = "/home/ubuntu/00HA1py/down/4dn/4DNFI3ULPBI5.pairs.gz"DATASET_DESC = "H1-hESC human embryonic stem cells Micro-C"DATASET_SIZE_GB = 29.38# Processing configuration - NO SKIPSSKIP_INITIAL = 0  # NO SKIP - process from beginningPAIRS_TO_READ = 800_000  # Read 800K consecutive pairs per roundSKIP_BETWEEN_ROUNDS = 0  # NO SKIP between roundsMAX_ROUNDS = 999999  # No limit on roundsMAX_PLOTS_TO_SHOW = 30  # Keep reasonable plot limit for display# Analysis parametersDIST_MIN = 8_000   # 8kb minimum distanceDIST_MAX = 40_000  # 40kb maximum distanceBIN_SIZE = 600     # 600bp binsMIN_COUNT = 5      # Minimum count thresholdWINDOW_RADIUS = 2  # ±2 bins for 5x5 windowRATIO_THRESHOLD = 1.20  # Center must be 20% higherMEDIAN_THRESHOLD = 0.40  # Median must be < 40% of center# Coldspot parameters - 2x20 checksCOLDSPOT_INITIAL_OFFSET = 0  # Start at 0 bins (adjacent)COLDSPOT_STEP = 5  # Step 5 bins diagonallyCOLDSPOT_CHECKS = 20  # Check 20 positions each direction (40 total)TMC_POWER = 1.3  # Power for top count in TMC# White-beige-black colormapwhite_beige_black = mcolors.LinearSegmentedColormap.from_list(    'white_beige_black',    [(1.0, 1.0, 1.0), (0.96, 0.87, 0.70), (0.0, 0.0, 0.0)])# Output directory with dataset namedef now_pdt():    return (datetime.now(timezone.utc) - timedelta(hours=7)).strftime("%Y-%m-%d %H:%M:%S PDT")today_tag = (datetime.now(timezone.utc) - timedelta(hours=7)).strftime("%y%m%d")OUT_DIR = f"/home/ubuntu/00HA1py/out/hotco/{today_tag}_(R441v6_Hotco_{DATASET_NAME}_{DATASET_CODENAME}_NoSkip)_results"os.makedirs(OUT_DIR, exist_ok=True)# Logging setupclass TeeOutput:    def __init__(self, name, mode):        self.file = open(name, mode)        self.stdout = sys.stdout        self.stderr = sys.stderr    def write(self, data):        self.stdout.write(data)        self.file.write(data)        self.file.flush()    def flush(self):        self.stdout.flush()        self.file.flush()logfile = os.path.join(OUT_DIR, f"{script_name}.log")sys.stdout = TeeOutput(logfile, 'a')  # append modesys.stderr = sys.stdout# Global trackingstart_time = time.time()plots_shown = 0cumulative_pairs = 0cumulative_unique_hotspots = set()cumulative_mb_processed = 0cumulative_valid_pairs = 0cumulative_chr_counts = defaultdict(int)current_chr_positions = {}round_yields = []print(f"{now_pdt()} {script_name} starting")print(f"Dataset: {DATASET_NAME} (codename: {DATASET_CODENAME}) - {DATASET_DESC}")print(f"Dataset size: {DATASET_SIZE_GB} GB")print(f"Input file: {DATASET_FILE}")print(f"FULL OUTPUT PATH: {OUT_DIR}/")print(f"  Main output: paired_hotcold_all.pkl")print(f"  Intermediate: paired_round###.pkl files")print(f"Config: 8-40kb range, 600bp bins, min 5 counts")print(f"Hotspot filter: 5x5 window, ratio>=1.20, median<40% of center")print(f"Coldspot: Start at 0 bins (adjacent), step by 5 bins, TMC metric")print(f"Coldspot checks: 2x20 positions (±0 to ±95 bins, ±0 to ±57kb)")print(f"Enforced 1:1 pairing - every hotspot gets exactly one coldspot")print(f"QC: Side-by-side worst hotspot / its coldspot (max {MAX_PLOTS_TO_SHOW} plots)")print(f"Processing: NO SKIPS - processing entire {DATASET_SIZE_GB}GB file from beginning")print("="*70)def get_folder_size_mb():    """Calculate total size of output folder in MB"""    total = 0    for dirpath, dirnames, filenames in os.walk(OUT_DIR):        for f in filenames:            fp = os.path.join(dirpath, f)            try:                total += os.path.getsize(fp)            except:                pass    return total / (1024 * 1024)def get_5x5_no_corners(matrix_5x5):    """Extract 21 values from 5x5 matrix, excluding 4 corners"""    values = []    for i in range(5):        for j in range(5):            if not ((i == 0 and j == 0) or (i == 0 and j == 4) or                     (i == 4 and j == 0) or (i == 4 and j == 4)):                values.append(matrix_5x5[i, j])    return np.array(values)def find_coldest_spot_tmc(contact_matrix, hotspot_bin1, hotspot_bin2):    """Find the coldest spot using TMC metric by checking 40 diagonal positions (2x20)"""    best_tmc = float('inf')    best_coldspot = None        # Check both diagonal directions: forward (+) and backward (-)    for direction in [1, -1]:        for step_num in range(COLDSPOT_CHECKS):  # 20 checks each direction            # Start at COLDSPOT_INITIAL_OFFSET, then add COLDSPOT_STEP for each check            offset = COLDSPOT_INITIAL_OFFSET + (step_num * COLDSPOT_STEP)            test_bin1 = hotspot_bin1 + (direction * offset)            test_bin2 = hotspot_bin2 + (direction * offset)                        # Skip if negative bins            if test_bin1 < 0 or test_bin2 < 0:                continue                        # Build 5x5 matrix for this position            matrix_5x5 = np.zeros((5, 5), dtype=float)            for di in range(-WINDOW_RADIUS, WINDOW_RADIUS + 1):                for dj in range(-WINDOW_RADIUS, WINDOW_RADIUS + 1):                    neighbor_bin = (test_bin1 + di, test_bin2 + dj)                    matrix_5x5[di + WINDOW_RADIUS, dj + WINDOW_RADIUS] = contact_matrix.get(neighbor_bin, 0)                        # Get values excluding corners            vals_no_corners = get_5x5_no_corners(matrix_5x5)                        if len(vals_no_corners) == 0:                continue                            max_val = np.max(vals_no_corners)            median_val = np.median(vals_no_corners)                        # Calculate TMC metric            tmc = (max_val ** TMC_POWER) * median_val                        # Keep track of best (lowest TMC)            if tmc < best_tmc:                best_tmc = tmc                                # Calculate ratio for consistency with hotspot format                sorted_vals = np.sort(vals_no_corners)                if len(sorted_vals) > 1 and sorted_vals[-2] > 0:                    ratio = sorted_vals[-1] / sorted_vals[-2]                else:                    ratio = float('inf') if sorted_vals[-1] > 0 else 0.0                                best_coldspot = ((test_bin1, test_bin2), max_val, ratio, matrix_5x5, median_val, tmc)        # If no coldspot found, create a default one at the first valid diagonal position    if best_coldspot is None:        # Try to find any valid position        for direction in [1, -1]:            test_bin1 = hotspot_bin1 + (direction * COLDSPOT_INITIAL_OFFSET)            test_bin2 = hotspot_bin2 + (direction * COLDSPOT_INITIAL_OFFSET)            if test_bin1 >= 0 and test_bin2 >= 0:                matrix_5x5 = np.zeros((5, 5), dtype=float)                best_coldspot = ((test_bin1, test_bin2), 0.0, 0.0, matrix_5x5, 0.0, 0.0)                break        return best_coldspotdef show_side_by_side_qc(hotspot_data, coldspot_data, round_pair_num, chr_name):    """Show side-by-side comparison of worst hotspot and its paired coldspot"""    global plots_shown        if plots_shown >= MAX_PLOTS_TO_SHOW:        return        # Unpack hotspot data    h_bin1, h_bin2, h_count, h_ratio, h_matrix, h_median = hotspot_data        # Unpack coldspot data    c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc = coldspot_data        print(f"\n  QC PLOT {plots_shown+1}/{MAX_PLOTS_TO_SHOW}: Rounds {round_pair_num*2-1}-{round_pair_num*2}")        # Create side-by-side plot    fig, axes = plt.subplots(1, 2, figsize=(14, 6))        # Plot hotspot (worst)    ax = axes[0]    non_zero_vals = h_matrix[h_matrix > 0]    if len(non_zero_vals) > 0:        h_vmax = np.percentile(h_matrix, 95)        if h_matrix[2, 2] > h_vmax * 1.5:            h_vmax = h_matrix[2, 2]    else:        h_vmax = 1        im = ax.imshow(h_matrix, cmap=white_beige_black, interpolation='nearest',                  vmin=0, vmax=h_vmax)    plt.colorbar(im, ax=ax, label='Contact count')        # Add text for hotspot    for i in range(5):        for j in range(5):            count_val = int(h_matrix[i, j])            is_corner = ((i == 0 and j == 0) or (i == 0 and j == 4) or                         (i == 4 and j == 0) or (i == 4 and j == 4))                        if count_val > 0 or is_corner:                text_color = 'white' if h_matrix[i, j] > h_vmax * 0.6 else 'black'                weight = 'bold' if (i == 2 and j == 2) else 'normal'                size = 10 if (i == 2 and j == 2) else 8                                if is_corner:                    text = f"X"                    ax.text(j, i, text, ha='center', va='center',                           color='red', fontsize=8, fontweight='normal')                else:                    ax.text(j, i, str(count_val), ha='center', va='center',                           color=text_color, fontsize=size, fontweight=weight)        h_pos1_mb = (h_bin1 * BIN_SIZE) / 1e6    h_pos2_mb = (h_bin2 * BIN_SIZE) / 1e6    h_distance = ((h_bin2 - h_bin1) * BIN_SIZE) / 1000        ax.set_title(f'WORST Hotspot: {chr_name} {h_pos1_mb:.1f}-{h_pos2_mb:.1f}Mb')    ax.set_xlabel(f'C={int(h_count)}, R={h_ratio:.2f}, Med={h_median:.1f}')        # Plot coldspot    ax = axes[1]    non_zero_vals = c_matrix[c_matrix > 0]    if len(non_zero_vals) > 0:        c_vmax = max(10, np.percentile(c_matrix, 95))  # At least 10 for visibility    else:        c_vmax = 10        im = ax.imshow(c_matrix, cmap=white_beige_black, interpolation='nearest',                  vmin=0, vmax=c_vmax)    plt.colorbar(im, ax=ax, label='Contact count')        # Add text for coldspot    for i in range(5):        for j in range(5):            count_val = int(c_matrix[i, j])            is_corner = ((i == 0 and j == 0) or (i == 0 and j == 4) or                         (i == 4 and j == 0) or (i == 4 and j == 4))                        if count_val > 0 or is_corner:                text_color = 'white' if c_matrix[i, j] > c_vmax * 0.6 else 'black'                weight = 'bold' if (i == 2 and j == 2) else 'normal'                size = 10 if (i == 2 and j == 2) else 8                                if is_corner:                    text = f"X"                    ax.text(j, i, text, ha='center', va='center',                           color='red', fontsize=8, fontweight='normal')                else:                    ax.text(j, i, str(count_val), ha='center', va='center',                           color=text_color, fontsize=size, fontweight=weight)        c_pos1_mb = (c_bin1 * BIN_SIZE) / 1e6    c_pos2_mb = (c_bin2 * BIN_SIZE) / 1e6        diagonal_offset = ((c_bin1 - h_bin1) * BIN_SIZE) / 1000    ax.set_title(f'Paired Coldspot: {c_pos1_mb:.1f}-{c_pos2_mb:.1f}Mb')    ax.set_xlabel(f'Max={c_max:.0f}, TMC={c_tmc:.1f}, Offset={diagonal_offset:+.1f}kb')        plt.suptitle(f'QC: Worst hotspot vs Its coldspot (D={h_distance:.1f}kb) - {DATASET_NAME} ({DATASET_CODENAME})')    plt.tight_layout()    plt.show()    plt.close()        plots_shown += 1def process_round(fh, round_num):    """Process one round of data collection - NO SKIPS VERSION"""    global cumulative_pairs, cumulative_unique_hotspots    global cumulative_mb_processed, cumulative_valid_pairs, cumulative_chr_counts    global current_chr_positions, round_yields        round_start = time.time()        # Data collection - per chromosome    contact_matrices = {}    lines_read = 0    cis_pairs = 0    valid_dist_pairs = 0        # Track positions for this round    round_chr_positions = {}        # Read PAIRS_TO_READ consecutive pairs    for i in range(PAIRS_TO_READ):        raw = fh.readline()        if not raw:            return None, None                lines_read += 1                    try:            parts = raw.decode('utf-8').strip().split('\t')            if len(parts) < 5:                continue                            chr1, pos1_str, chr2, pos2_str = parts[1:5]                        if not chr1.startswith('chr'):                chr1 = 'chr' + chr1            if not chr2.startswith('chr'):                chr2 = 'chr' + chr2                        if chr1 != chr2:                continue                        cis_pairs += 1            cumulative_chr_counts[chr1] += 1            pos1 = int(pos1_str)            pos2 = int(pos2_str)                        if chr1 not in round_chr_positions:                round_chr_positions[chr1] = [float('inf'), 0]            round_chr_positions[chr1][0] = min(round_chr_positions[chr1][0], pos1, pos2)            round_chr_positions[chr1][1] = max(round_chr_positions[chr1][1], pos1, pos2)                        if pos1 > pos2:                pos1, pos2 = pos2, pos1                        dist = pos2 - pos1            if dist < DIST_MIN or dist > DIST_MAX:                continue                        valid_dist_pairs += 1                        if chr1 not in contact_matrices:                contact_matrices[chr1] = defaultdict(int)                        bin1 = pos1 // BIN_SIZE            bin2 = pos2 // BIN_SIZE                        contact_matrices[chr1][(bin1, bin2)] += 1                    except Exception as e:            continue        # Calculate MB processed and update positions    round_mb = 0    current_chr_str = ""    for chr_name, (min_pos, max_pos) in round_chr_positions.items():        mb_span = (max_pos - min_pos) / 1e6        round_mb += mb_span        current_chr_positions[chr_name] = max_pos / 1e6        if round_chr_positions:        max_chr = max(round_chr_positions.keys(),                      key=lambda x: cumulative_chr_counts[x] if x in cumulative_chr_counts else 0)        current_chr_str = f"{max_chr}:{current_chr_positions[max_chr]:.1f}Mb"        cumulative_mb_processed += round_mb    cumulative_valid_pairs += valid_dist_pairs        # Process hotspots and find paired coldspots    all_round_pairs = []    worst_hotspot_for_qc = None        for chr_name, contact_matrix in contact_matrices.items():        if len(contact_matrix) < 10:            continue                # Find hotspots first        high_count_bins = [(k, v) for k, v in contact_matrix.items() if v >= MIN_COUNT]                for (bin1, bin2), center_count in high_count_bins:            # Build 5x5 matrix            matrix_5x5 = np.zeros((5, 5), dtype=float)            for di in range(-WINDOW_RADIUS, WINDOW_RADIUS + 1):                for dj in range(-WINDOW_RADIUS, WINDOW_RADIUS + 1):                    neighbor_bin = (bin1 + di, bin2 + dj)                    matrix_5x5[di + WINDOW_RADIUS, dj + WINDOW_RADIUS] = contact_matrix.get(neighbor_bin, 0)                        center_val = matrix_5x5[2, 2]            vals_no_corners = get_5x5_no_corners(matrix_5x5)                        # Check if this is a hotspot            if np.count_nonzero(vals_no_corners) <= 1:                ratio = float('inf') if center_val > 0 else 0.0                median_val = 0                is_hotspot = ratio >= RATIO_THRESHOLD            else:                median_val = np.median(vals_no_corners)                                if center_val > 0 and median_val >= center_val * MEDIAN_THRESHOLD:                    continue                                sorted_vals = np.sort(vals_no_corners)                top1 = sorted_vals[-1]                top2 = sorted_vals[-2]                                if center_val < top1:                    continue                                if top2 > 0:                    ratio = center_val / top2                else:                    ratio = float('inf')                                is_hotspot = ratio >= RATIO_THRESHOLD                        if is_hotspot:                cumulative_unique_hotspots.add((chr_name, bin1, bin2))                                # Find the coldest spot for this hotspot using TMC                coldspot = find_coldest_spot_tmc(contact_matrix, bin1, bin2)                                if coldspot:                    (c_bin1, c_bin2), c_max, c_ratio, c_matrix, c_median, c_tmc = coldspot                                        # Create paired entry                    paired_entry = ((chr_name, bin1, bin2), center_count, ratio,                                  (c_bin1, c_bin2), c_max, c_ratio)                    all_round_pairs.append(paired_entry)                                        # Track worst hotspot for QC                    if worst_hotspot_for_qc is None or ratio < worst_hotspot_for_qc[3]:                        worst_hotspot_for_qc = (bin1, bin2, center_count, ratio, matrix_5x5, median_val,                                               c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc,                                               chr_name)        cumulative_pairs += len(all_round_pairs)    round_yields.append(len(all_round_pairs))        # Save round results    if all_round_pairs:        save_file = os.path.join(OUT_DIR, f"paired_round{round_num:03d}.pkl")        with open(save_file, 'wb') as f:            pickle.dump(all_round_pairs, f)        # Print summary    elapsed_total = (time.time() - start_time) / 3600    folder_mb = get_folder_size_mb()    median_yield = np.median(round_yields) if round_yields else 0        print(f"R{round_num:03d}: +{len(all_round_pairs):4d} pairs | Total: {cumulative_pairs:6,} | {current_chr_str}")    print(f"     {elapsed_total:5.2f}h | {cumulative_mb_processed:7.1f}Mb | {folder_mb:5.1f}MB | Med: {median_yield:.0f}/round")        return all_round_pairs, worst_hotspot_for_qcdef main():    try:        # Check input file exists        if not os.path.exists(DATASET_FILE):            print(f"ERROR: Input file not found: {DATASET_FILE}")            print(f"\n{script_name} terminated - missing input file")            raise FileNotFoundError(f"Input file not found: {DATASET_FILE}")                print(f"Input file size: {os.path.getsize(DATASET_FILE) / (1024**3):.2f} GB")                all_pairs_list = []        round_num = 0        round_pair_num = 0        two_round_worst = None                with gzip.open(DATASET_FILE, 'rb') as fh:            # NO SKIPPING - process from beginning            print(f"Processing from beginning - NO SKIPS for {DATASET_NAME}")            print("="*70)                        while round_num < MAX_ROUNDS:                round_num += 1                                # Process one round                round_pairs, worst_hotspot = process_round(fh, round_num)                if round_pairs is None:                    print("\nEnd of file reached")                    break                                if round_pairs:                    all_pairs_list.extend(round_pairs)                                # Track worst hotspot across 2 rounds                if worst_hotspot and (two_round_worst is None or worst_hotspot[3] < two_round_worst[3]):                    two_round_worst = worst_hotspot                                # Every 2 rounds, show side-by-side QC                if round_num % 2 == 0:                    round_pair_num += 1                    if plots_shown < MAX_PLOTS_TO_SHOW and two_round_worst:                        # Unpack the worst hotspot data                        (h_bin1, h_bin2, h_count, h_ratio, h_matrix, h_median,                         c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc,                         chr_name) = two_round_worst                                                hotspot_data = (h_bin1, h_bin2, h_count, h_ratio, h_matrix, h_median)                        coldspot_data = (c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc)                                                show_side_by_side_qc(hotspot_data, coldspot_data, round_pair_num, chr_name)                    two_round_worst = None                                # NO SKIP between rounds - continue immediately                # Process last odd round if exists        if two_round_worst and plots_shown < MAX_PLOTS_TO_SHOW:            round_pair_num += 1            (h_bin1, h_bin2, h_count, h_ratio, h_matrix, h_median,             c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc,             chr_name) = two_round_worst                        hotspot_data = (h_bin1, h_bin2, h_count, h_ratio, h_matrix, h_median)            coldspot_data = (c_bin1, c_bin2, c_max, c_ratio, c_matrix, c_median, c_tmc)                        show_side_by_side_qc(hotspot_data, coldspot_data, round_pair_num, chr_name)                # Final summary        print("="*70)        print(f"ANALYSIS COMPLETE at {now_pdt()}")        print(f"Dataset: {DATASET_NAME} (codename: {DATASET_CODENAME})")        print(f"Runtime: {(time.time() - start_time)/3600:.2f} hours")        print(f"Rounds: {round_num}, Genomic span: ~{cumulative_mb_processed:.1f}Mb")        print(f"Total paired entries: {cumulative_pairs:,}")        print(f"Unique hotspots: {len(cumulative_unique_hotspots):,}")                # Save final paired data        final_file = os.path.join(OUT_DIR, "paired_hotcold_all.pkl")        seen = {}        for entry in all_pairs_list:            (chr_name, h_bin1, h_bin2), h_count, h_ratio, (c_bin1, c_bin2), c_max, c_ratio = entry            key = (chr_name, h_bin1, h_bin2)            # Keep the entry with highest hotspot ratio            if key not in seen or h_ratio > seen[key][2]:                seen[key] = entry                final_data = list(seen.values())        with open(final_file, 'wb') as f:            pickle.dump(final_data, f)                print(f"\nOUTPUT FILE:")        print(f"  Paired data: {final_file} ({len(final_data):,} unique pairs)")        print(f"  Format: ((chr, h_bin1, h_bin2), h_count, h_ratio, (c_bin1, c_bin2), c_max, c_ratio)")        print(f"Folder size: {get_folder_size_mb():.1f} MB")            except Exception as e:        print(f"Error: {e}")        import traceback        traceback.print_exc()if __name__ == "__main__":    main()print(f"\n{script_name} complete")#chm3
```
